```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(name,n,m,i,title):
    plt.subplot(n,m,i)
    plt.imshow(name,cmap='gray')
    plt.title(title)
    plt.axis('off')
```

```python
def low_pass_filter(image, kernel_size):
    return cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)
```

```python
def high_pass_filter(image, kernel_size):
    low_pass = low_pass_filter(image, kernel_size)
    return cv2.subtract(image, low_pass)
```

```python
img1_path = 'modi.jpeg'
img2_path = 'rahul.jpg'
img1 = cv2.imread(img1_path, cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread(img2_path, cv2.IMREAD_GRAYSCALE)
img1_resized = cv2.resize(img1, (512, 512))
img2_resized = cv2.resize(img2, (512, 512))
kernel_size = 31
high_pass_img = high_pass_filter(img1_resized, kernel_size)
low_pass_img = low_pass_filter(img2_resized, kernel_size)
hybrid_img = cv2.addWeighted(high_pass_img, 0.5, low_pass_img, 0.5, 0)
```

```python
plt.figure(figsize=(20, 8))
show(img1_resized, 1, 5, 1, "Original Image 1")
show(img2_resized, 1, 5, 2, "Original Image 2")
show(high_pass_img, 1, 5, 3, "High-Pass Filtered")
show(low_pass_img, 1, 5, 4, "Low-Pass Filtered")
show(hybrid_img, 1, 5, 5, "Hybrid Image")

plt.tight_layout()
plt.show()
```
